```python
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize
```

```python
jokes = []
count = 0
while count < 10:
    res = requests.get('https://api.chucknorris.io/jokes/random')
    chuck = res.json()
    jokes.append(chuck)
    count = count + 1
    
jokes
```

```python
flattened_data = json_normalize(jokes)
```

```python
df2 = pd.DataFrame(flattened_data)
df2
```

```python
jokes = df2['value']
```

```python
jokes.head()
```

```python
df2.to_excel('./chuck.xls', index=False)
```

# API2

```python
response = requests.get('https://api.punkapi.com/v2/beers')
```

```python
beer = response.json()
beer
```

```python
beer2 = json_normalize(beer)

beer2.head(5)
```

```python
beer2.columns
```

```python
beer = beer2[['id', 'name', 'tagline', 'first_brewed', 'description','abv', 'ibu','volume.unit']]
```

```python
beer.head()
```

```python
beerjokes = pd.merge(beer, jokes, left_index=True, right_index=True) # A Chuck Norris Joke for every beer! :)
```

```python
beerjokes.rename(columns={'value': 'joke'}, inplace=True)
```
